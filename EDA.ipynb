```python
sales = pd.read_csv('/kaggle/input/video-games-sales-2019/vgsales-12-4-2019.csv')
sales.head(10)
```

```python
sales.columns
```

```python
sales = sales[['Name','Genre', 'Platform',
       'Publisher', 'Critic_Score',
       'Total_Shipped', 'Global_Sales', 'Year', 'Last_Update', 'url','img_url']]
sales.head(10)
```

```python
sales_by_genre = sales.groupby('Genre')['Global_Sales'].sum()
sales_by_genre
```

```python
sales_by_genre.index
```

```python
list(sales_by_genre.values)
```

```python
sports_games = sales.loc[sales['Genre'] == 'Sports']
total_sales = np.sum(sports_games['Global_Sales'])
sports_games['Sales_pct'] = sports_games['Global_Sales'].apply(lambda x: (x / total_sales) * 100)
sports_games = sports_games.sort_values(by=['Global_Sales'], ascending=False)
sports_games = sports_games.groupby('Name')['Global_Sales'].sum()
#sports_games[['Name','Global_Sales']]['Global_Sales'].values[:10]
sports_games.sort_values(ascending=False).values[:10]
```

```python
sports_games_short = pd.DataFrame({'Name': list(sports_games.sort_values(ascending=False).index), 'Global_Sales': sports_games.sort_values(ascending=False).values})
sports_games_short['temp1'] = sports_games_short['Name'].apply(lambda x: x[:4])
fifa_games = sports_games_short.loc[sports_games_short['temp1'] == 'FIFA']
fifa = fifa_games.reset_index()[['Name','Global_Sales']]
```

```python
total_fifa_sales = np.sum(fifa['Global_Sales'])
fifa['pct'] = fifa['Global_Sales'].apply(lambda x: (x/total_fifa_sales)*100)
fifa_topten = fifa.iloc[:10,:]
fifa_topten
```

```python
cs = sales.groupby('Name')['Critic_Score'].max().values
gs = sales.groupby('Name')['Global_Sales'].sum().values
yr = sales.groupby('Name')['Year'].max().values
csgs = pd.DataFrame({'Name': list(sales.groupby('Name')['Critic_Score'].max().index), 'Critic_Score':cs,'Global_Sales':gs, 'Year':yr})
csgs = csgs.dropna()
csgs = csgs.loc[csgs['Global_Sales'] > 0]
csgs
```

```python
csgs[['Critic_Score', 'Global_Sales']].to_csv('csgs.csv',index=False)
```

```python
coors = []
for i in range(len(gs)):
    if np.isnan(gs[i]) == False and np.isnan(cs[i]) == False and gs[i] > 0:
        coors.append([cs[i],gs[i]])
```
